```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('eleitorado_municipio_2018.csv', sep = ';', encoding = 'cp1252')
df.head()
```

```python
#verificando municipios unicos
df['CD_MUNICIPIO'].nunique()
```

```python
# verificando municipios fora do país
df[df['SG_UF'] == 'ZZ'].sample(4).T
```

```python
# verificando quantos municipios há fora do país
df[df['SG_UF'] == 'ZZ']['CD_MUNICIPIO'].nunique()
```

```python
# excluindo do df os municipios fora do país
df = df[df['SG_UF'] != 'ZZ']
```

```python
# verificando o total de municipios brasileiros
df['CD_MUNICIPIO'].nunique()
```

```python
# quantidade de eleitores com 19 anos
df['QTD_ELEITORES_19'].sum()
```

```python
print('Eleitorado brasileiro')
print('Mulheres:' , df['QTD_ELEITORES_FEMININO'].sum())
print('Homens:' , df['QTD_ELEITORES_MASCULINO'].sum())
print('Deficientes:' , df['QTD_ELEITORES_DEFICIENTE'].sum())
print('Não informado:' , df['QTD_ELEITORES_NAOINFORMADO'].sum())
```

```python
# criando variáveis para calcular porcentagem
tot_eleitores = df['QTD_ELEITORES'].sum()
tot_fem = df['QTD_ELEITORES_FEMININO'].sum()
tot_masc = df['QTD_ELEITORES_MASCULINO'].sum()
tot_nao = df['QTD_ELEITORES_NAOINFORMADO'].sum()
```

```python
print('Mulheres:' , ((tot_fem / tot_eleitores) * 100) .round(2))
print('Homens:' , ((tot_masc / tot_eleitores) * 100) .round(2))
print('Não informado:' , ((tot_nao / tot_eleitores) * 100) .round(2))
```

```python
df[df['QTD_ELEITORES_MASCULINO'] > df['QTD_ELEITORES_FEMININO']].count()
```

```python
# criando uma lista e somando os valores
lista = ['QTD_ELEITORES_FEMININO','QTD_ELEITORES_MASCULINO','QTD_ELEITORES_DEFICIENTE']
tot_eleitores = df[lista].sum()
```

```python
df.groupby('QTD_ELEITORES_FEMININO')['NM_MUNICIPIO'].max()
```

```python
# 5 locais com o maior numero de mulheres
df.nlargest(5,'QTD_ELEITORES_FEMININO')
```

```python
# Criando coluna para ralação masculino e feminino
df['RELACAO_FM'] = df['QTD_ELEITORES_FEMININO'] / df['QTD_ELEITORES_MASCULINO']
```

```python
# 5 locais com a maior relação m/f
df.nlargest(5,'RELACAO_FM').T
```

```python
# 5 locais com a maior quantidade de eleitores com 100 anos ou mais
df.nlargest(5,'QTD_ELEITORES_MAIORIGUAL100').T
```

```python
# 5 locais com a maior quantidade de eleitores deficientes
df.nlargest(5,'QTD_ELEITORES_DEFICIENTE').T
```

```python
# gráfico de relação m/m
df['RELACAO_FM'].plot.hist(bins = 100)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# plotar a distribuição da ralaçao f/m
sns.distplot(df['RELACAO_FM'], bins = 100, color = 'red', kde = False)

#embelezando o grafico
plt.title('Relação Eleitoras/Eleitores', fontsize = 18)
plt.xlabel('Eleitoras/Eleitores', fontsize = 14)
plt.ylabel('Frequencia', fontsize = 14)
plt.axvline(1.0, color = 'black', linestyle='--')
```
