# Reembolso dos senadores no ano de 2018

```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('2018.csv',encoding = 'latin1',sep = ';',decimal = ',',skiprows = 1)
df.head()
```

```python
# soma dos valores reembolsados
df['VALOR_REEMBOLSADO'].sum()
```

```python
# Quantas vezes cada senador pediu reembolso
df['SENADOR'].value_counts()
```

```python
# total do valor reembolsado por cada senador em ordem decrescente
df.groupby('SENADOR')['VALOR_REEMBOLSADO'].sum().sort_values(ascending = False)
```

```python
3 5 maiores valores reembolsados
df.nlargest(5,'VALOR_REEMBOLSADO')
```

```python
# 5 menores valores reembolsados
df.nsmallest(5, 'VALOR_REEMBOLSADO')
```

```python
df['TIPO_DESPESA'].value_counts()
```

```python
# total do valor reembolsado por tipo de despesa de cada senador
df.groupby(['SENADOR','TIPO_DESPESA']).sum()
```
