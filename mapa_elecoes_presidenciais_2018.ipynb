```python
# vamos carregar o arquivo com o resultado da eleição brasil dividido por municipio/zona
df_votacao = pd.read_csv('./dados_eleicao/votacao_candidato_munzona_2018_BR.csv', sep = ';', encoding = 'latin1')

df_votacao.head()
```

```python
# visializando o dataframe transposto
df_votacao.head().T
```

```python
# separar os dados pois precidamos apenas dos dados do segundo turno
df_votacao = df_votacao[df_votacao['NR_TURNO'] == 2]

df_votacao.sample(3).T
```

```python
# verificar quantos municipios unicos existem nos dados
# vamso pegar por código do municipio
df_votacao['CD_MUNICIPIO'].nunique()
```

```python
# brasil tem 5570 municipios
# vamos verificar as cidades fora do brasil
df_votacao[df_votacao['SG_UF'] == 'ZZ'].sample(4).T
```

```python
# quantos municipios votantes há fora do país
df_votacao[df_votacao['SG_UF'] == 'ZZ']['CD_MUNICIPIO'].nunique()
```

```python
#vamos eliminar os municipios de fora do pais
df_votacao = df_votacao[df_votacao['SG_UF'] != 'ZZ']
```

```python
# verificando a quantidade de municipios
df_votacao['CD_MUNICIPIO'].nunique()
```

```python
# OS DADOS ESTAO SEPARADOS POR MUNICIPIO/ZONA
# municipios pequenos normalmente tem apenas 1 zona eleitoral
# mas municipios maiores podem ter várias zonas eleitorais

# vamos ferificar quantas zonas ha em belo horizonte
df_votacao[df_votacao['NM_MUNICIPIO'] == 'BELO HORIZONTE']
```

```python
# escolhendo as colunas a serem visualizadas
df_votacao = df_votacao[['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO', 'QT_VOTOS_NOMINAIS']]

df_votacao.head()
```

```python
# precisamos agregar dados por municipio a cada dado
df_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO']).sum()
```

```python
# fazer o groupby, reindexar o indice, vai salvar isso em um dataframe
df_votacao = df_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO']).sum().reset_index()

df_votacao.sample(4)
```

```python
# vamos verificar novamente belo horizonte
df_votacao[df_votacao['NM_MUNICIPIO'] == 'BELO HORIZONTE']
```

```python
# ordenar o datafreme pela quantidade de votos nominais e deixar apenas o primeiro registro de cada municipio
# registro de cada municipio
df_votacao.sort_values(by = 'QT_VOTOS_NOMINAIS', ascending = False).head(10)
```

```python
#  vamos ordenar 
df_votacao.sort_values(by = 'QT_VOTOS_NOMINAIS', ascending = False, inplace = True)

#dropar as linhas duplicadas mantendo a primeira
df_votacao.drop_duplicates(subset = 'CD_MUNICIPIO', keep = 'first', inplace = True)
```

```python
df_votacao.head(10)
```

```python
# vamos verificar novamente belo horizonte
df_votacao[df_votacao['NM_MUNICIPIO'] == 'BELO HORIZONTE']
```

```python
# vamos criar coluna representando o resultado da eleição no municipio
# sendo que o resultado é true se o candidato bolsonaro venceu 
# e false em caso contrario

df_votacao['RESULTADO'] = df_votacao['NR_CANDIDATO'].apply(lambda x: x == 17)
```

```python
# verificando o resultado
df_votacao.head(10)
```

```python
# mostrando valores do dataframe geopandas
df_mapa.sample(3)
```

```python
# tentando encrontrar o municipio pleo codigo
df_votacao[df_votacao['CD_MUNICIPIO'] == 3147204]
```

```python
# pq nao podemos usar o nome do municipio para juntas os dados
# pode haver varias cidades com o mesmo nome
df_votacao[df_votacao['NM_MUNICIPIO'] == 'OURO BRANCO']
```

```python
#RECUPERAR A TEBELA DE CORRESPONDENCIA DO TSE DIRETO DO GITHUB DO ESTADAO
link = 'https://raw.githubusercontent.com/estadao/como-votou-sua-vizinhanca/master/data/votos/correspondencia-tse-ibge.csv'

df_equivalencia = pd.read_csv(link)
df_equivalencia.head()
```

```python
# vamos prepar  os dataframes
# reindexar o datafreme de votação pelo codigo de tse
df_vot_novo = df_votacao.copy()

df_vot_novo.set_index('CD_MUNICIPIO', drop = False, inplace = True)

df_vot_novo.head()
```

```python
# vamos reindexar o datafreme de equivalencia pelo codigo do tse
df_equi_novo = df_equivalencia.copy()

df_equi_novo.set_index('COD_TSE', drop = False, inplace = True)

df_equi_novo.head()
```

```python
#verificando o tamanho dos dataframes
df_vot_novo.info()
```

```python
df_equi_novo.info()
```

```python
#criando um novo datafreme com base no merge de votação e equivalencia
df_vot_equi = df_vot_novo.join(df_equi_novo)
df_vot_equi.head()
```

```python
# vamos juntar o dataframe do geopandas com o df votação/equivalencia

df_vot_equi.info()
```

```python
df_mapa.info()
```

```python
# transformar a coluna do codigo do ibge do dataframe de votação/equivalencia
df_vot_equi['GEOCOD_IBGE'] = df_vot_equi['GEOCOD_IBGE'].astype(str)

df_vot_equi.info()
```

```python
# vamos unificar os indices
#mudando o indice do datafreme do mapa
df_mapa.set_index('CD_GEOCMU', drop = False, inplace = True)
```

```python
#mudando o indice do datafreme de votacao/equivalencia
df_vot_equi.set_index('GEOCOD_IBGE', drop = False, inplace = True)
```

```python
# vamos juntar os datafremaes
df_mapa_novo = df_mapa.join(df_vot_equi)

df_mapa_novo.head().T
```

```python
# tamanho do mapa novo
df_mapa_novo.info()
```

```python
# vamos descobrir os 2 municipios a mais no mapa
df_mapa_novo[df_mapa_novo['NR_CANDIDATO'].isnull()]
```

```python
# matar a coluna de ajuste
df_mapa_novo.drop(columns=['AJUSTE'], inplace = True)
```

```python
# vamos eliminar as lagoas 
df_mapa_novo.dropna(inplace = True)
```

```python
df_mapa_novo.info()
```

```python
# vamos desenhar o mapa
# criar a referencia para o codigo de cor
# sendo que false/0 - red, True/1-green
from matplotlib.colors import ListedColormap

cmap = ListedColormap(['red', 'green'])


#desenhando o mapa
fig, ax = plt.subplots(1, figsize = (12,12))
ax = df_mapa_novo.plot(column='RESULTADO', cmap = cmap, legend = True, linewidth = 0.1, ax = ax, 
                       edgecolor='grey')

#esconder os eixos
ax.set_axis_off()
#colocando titulo
ax.set_title('Resuldato Eleição 2018 por Município / 2° Turno', fontdict={'fontsize': '25', 'fontweight':'3'})

#criar nota de rodape
    
ax.annotate('Fonte: IBGE, 2015, TSE, 2019; JORNAL ESTADO DE SAO PAULO 2019', xy = (0.1, 0.08), xycoords = 'figure fraction', horizontalalignment = 'left', verticalalignment = 'top', fontsize = 10, color = '#555555')
# alterando a legenda
leg = ax.get_legend()
leg.get_texts()[0].set_text('Fernando Haddad')
leg.get_texts()[1].set_text('Jair Bolsonaro')

plt.show()
```
